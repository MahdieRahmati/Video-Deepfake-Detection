# finding paths

```python
def get_fake_paths(main_directory):
    data_frames = []
    person_label = 0 
    
    for directory in Path(main_directory).glob('*'):

        if directory.is_dir(): #persons 
            for subdir in directory.glob('*'): #videos of person
                if subdir.is_dir():
                    data = []
                    person_name = subdir.parts[-2] 
                    image_files = list(subdir.glob('*.jpg')) 
                    image_files = sorted(image_files, key=lambda x: x.name)
                    for image_file in image_files:
                        data.append({'Image_Path': str(image_file), 'Person_Name': person_name, 'Person_Label': person_label})
                    df = pd.DataFrame(data)
                    data_frames.append(df)
        
            person_label = person_label + 1         
    return data_frames
```

```python
high_directory = "/kaggle/input/deepfaketimit/DeepfakeTIMIT/higher_quality"
low_directory = "/kaggle/input/deepfaketimit/DeepfakeTIMIT/lower_quality"

high_df = get_fake_paths(high_directory)
low_df = get_fake_paths(low_directory)
```

```python
def get_real_paths(main_directory):
    data_frames = []
    person_label = 32 
    excluded_persons = ['fcmr0', 'fdms0', 'mabw0', 'fpkt0', 'fgjd0', 'mtas1', 'mtmr0', 'mbjk0', 'mbdg0', 'fcrh0', 'mreb0']
    excluded_videos = ["head", "head2", "head3"]


    for directory in Path(main_directory).glob('*'):
        if directory.is_dir() and directory.parts[-1] not in excluded_persons: #persons 
            directory = directory / directory.parts[-1] / "video"
            for subdir in directory.glob('*'): #videos of person
        
                if subdir.is_dir() and subdir.parts[-1] not in excluded_videos:
                    data = []
                    person_name = subdir.parts[-3]  
                    image_files = list(subdir.glob('*')) 
                    image_files = sorted(image_files, key=lambda x: x.name)
                    for image_file in image_files:
                        data.append({'Image_Path': str(image_file), 'Person_Name': person_name ,'Person_Label': person_label})
                    df = pd.DataFrame(data)
                    data_frames.append(df)
            
            person_label = person_label+1
            print(person_label)

    return data_frames
```

```python
import cv2
import dlib
import pandas as pd

def detect_faces_and_save_coordinates(df, detector):
    for i in range(len(df)):
        
        for index, row in df[i].iterrows():
            image_path = row['Image_Path']
            frame = cv2.imread(image_path)
            face_rects, _, _ = detector.run(frame, 0)
            if not face_rects:
                print("No face detected in:", image_path)
                df[i].at[index, 'x1'] = None
                df[i].at[index, 'x2'] = None
                df[i].at[index, 'y1'] = None
                df[i].at[index, 'y2'] = None
            else:
                x1, y1, x2, y2 = face_rects[0].left(), face_rects[0].top(), face_rects[0].right(), face_rects[0].bottom()
                df[i].at[index, 'x1'] = x1
                df[i].at[index, 'x2'] = x2
                df[i].at[index, 'y1'] = y1
                df[i].at[index, 'y2'] = y2
    return df
```

```python
detector = dlib.get_frontal_face_detector()
high_df_with_coordinates = detect_faces_and_save_coordinates(high_df, detector)
```

```python
low_df_with_coordinates = detect_faces_and_save_coordinates(low_df, detector)
```
